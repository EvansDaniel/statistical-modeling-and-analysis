```r
library(Stat2Data)
data("Blood1")
attach(Blood1)

#Graphical Summaries
par(mfrow=c(1,1))
stripchart(SystolicBP, method="stack", pch=19, 
           main="Dotplot of Systolic BP", offset=0.5,
           xlab="Blood Pressure")
stripchart(SystolicBP~Overwt, method="stack", pch=19, 
           main="Comparing Systolic BP over Weight", offset=0.5,
           xlab="Blood Pressure")

#Numerical Summaries
n = tapply(SystolicBP, Overwt,  length)
mean= round(tapply(SystolicBP, Overwt, mean),2)
sd = round(tapply(SystolicBP, Overwt, sd),2)
cbind(n, mean, sd)

n.tot = length(SystolicBP)
mean.tot = mean(SystolicBP)
sd.tot = sd(SystolicBP)
cbind(n.tot, mean.tot, sd.tot)

#Select 3 adults at random
x1 = sample(1:500, 3, replace = TRUE)

#Identify explanatory variable as a categorical variable
weight = as.factor(Overwt)

#Define the model
blood.lm=lm(SystolicBP~weight)
blood.lm2=lm(SystolicBP~Overwt)
#anova(blood.lm)
#anova(blood.lm2)

blood.aov=aov(SystolicBP~weight)
summary(blood.aov)

#Assessing Model Conditions
par(mfrow = c(2,2))
hist(blood.lm$residuals, main=NULL)
plot(blood.lm$fitted.values, blood.lm$residuals)
abline(h=0)
qqnorm(blood.lm$residuals, main=NULL)
qqline(blood.lm$residuals)

s.min = min(sd)
s.max = max(sd)
s.max/s.min

#Fisher's LSD
library(agricolae)
fish = LSD.test(SystolicBP, weight, 497, 363286)
fish

#Alternative adjustments to control for Type I Error
TukeyHSD(blood.aov)
plot(TukeyHSD(blood.aov))

# Pairwise T-tests
pairwise.t.test(SystolicBP, weight)  #Fisher LSD
pairwise.t.test(SystolicBP, weight, p.adj="bonferroni")    #Bonferroni

#Plot means individually
#library(gplots)
#plotmeans(SystolicBP~weight*Smoke)

#Two-way ANOVA
weight = as.factor(Overwt)
smoke = as.factor(Smoke)

library(ggplot2)
ggplot(Blood1, aes(x=Smoke, y=SystolicBP, fill=weight))+geom_boxplot()

interaction.plot(Smoke, weight, SystolicBP)

blood.two=aov(SystolicBP~Smoke*weight)
coef(blood.two)
summary(blood.two)
plot(blood.two)

#Investigating simple effects
library(lsmeans)
comp=lsmeans(blood.two, pairwise~Smoke:weight, adjust="tukey")
comp

library(multcompView)
cld(comp, alpha=0.05, Letters=letters, adjust = "tukey")
```
