```r
library(Stat2Data)
library(lattice)
data(Jurors)
head(Jurors)
```

```r
hist(Jurors$Year)
```

```r
hist(Jurors$Period)
```

```r
xyplot(Jurors$PctReport ~ Jurors$Period, group=Jurors$Year, type=c('r','p'), auto.key=TRUE)
print(cor(Jurors$PctReport, Jurors$Period))
```

```r
summary(Jurors)
```

```r
periodI2000Lm = lm(Jurors$PctReport ~ Jurors$Period + Jurors$I2000)
summary(periodI2000Lm)
```
