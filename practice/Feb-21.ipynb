```r
library(Stat2Data)
library(lattice)
data(Perch)
head(Perch)
```

```r
pairs(~ Perch$Weight + Perch$Width + Perch$Length)
```

```r
Perch.lm1 <- lm(Weight ~ Length + Width, data = Perch)
summary(Perch.lm1)
```

```r
Perch.lm2 <- lm(Weight ~ Length*Width, data = Perch)
summary(Perch.lm2)
```

```r
par(mfrow=c(2,2))
plot(Perch.lm1)
```

```r
par(mfrow=c(2,2))
plot(Perch.lm2)
```

```r
Perch.lm2 <- lm(Weight ~ Length + Width + I(Length^2), data = Perch)
summary(Perch.lm2)
```

```r
Perch.lm3 <- lm(Weight ~ Length + Width + I(Length^2), data = Perch)
summary(Perch.lm3)
```

```r
par(mfrow=c(2,2))
plot(Perch.lm3)
```
