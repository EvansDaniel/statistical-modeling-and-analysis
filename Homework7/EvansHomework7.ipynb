```r
library(readr)

air = read_csv('Paper Airplane Data 214.csv')
head(air)
attach(air)
```

### Theoretical Model

```r
#Two-way ANOVA
air.aov=aov(Length~Design*Hand)
summary(air.aov)
```

```r
plot(air.aov)
```

### The model indicates that the distance an airplane goes depends on its design. However, this is not the case with hand variable, which is a somewhat unexpected result. The interaction plot below shows that the dominant hand performs better than the nondominant hand in all but one case. According to the p-value, this is due only to random chance, but perhaps another study should be done to verify these results and ensure a type II error was not committed. Furthermore, the model suggests that there is no interaction b/w the explanatory variables. Thus, the distance an airplane is thrown  given a certain design is not significantly different depending on whether it was thrown with the dominant or nondominant hand. This can be seen in the interaction plot because the lines are roughly parallel.

### We have very good looking normality based on the QQ plot, but the residuals vs fitted graph indicates that there is non-constant variance in the residuals. The way in which we collected the data ensures that there is approximately zero mean and that the data points are independent. 

```r
interaction.plot(Design, Hand, Length)
```
